```python
from pylab import plot, draw, pause
from numpy import arange, sin, pi
```

```python
x = arange(0, 2 * pi, 0.01)
```

```python
line, = plot(x, sin(x))
for i in arange(1, 200):
    line.set_ydata(sin(x + i / 10.0))
    draw()
    pause(0.0001)
```
